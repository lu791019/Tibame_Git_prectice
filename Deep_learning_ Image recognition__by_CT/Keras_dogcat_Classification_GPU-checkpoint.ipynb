# transfering learning 遷移學習


```python
from glob import glob
import pandas as pd
dirname = 'C:/Users/Big data/Keras_CNN_class/train'

df_cat = pd.DataFrame()
df_cat['path'] = glob('train/cat*.*')
df_cat['target'] = 0
df_cat
```

```python
df_dog = pd.DataFrame()
df_dog['path'] = glob('train/dog*.*')
df_dog['target'] = 1
df_dog 
```

```python
df = pd.concat([df_cat,df_dog],axis=0,ignore_index=True)
```

```python
from keras.applications.vgg16 import VGG16

#inclued_top = False --> 不通用原本的MLP,因為現在只用辨認兩種
#input_shape 將圖片壓縮成相同大小
vgg = VGG16(include_top=False,input_shape=(224,224,3))
```

```python
vgg.summary()
```

```python
from keras.layers import Dense,Flatten,Dropout
from keras.models import Model
```

```python
vgg.layers
```

```python
for l in vgg.layers:
    l.trainable = False
x = Flatten()(vgg.output)
#平坦化後創建新的MLP 以DENSE新增神經元節點; 以DROPOUT捨棄一定程度量(%)的神經元以避免過擬合
x = Dense(2048,activation='relu')(x)
x = Dropout(0.25)(x)
x = Dense(512,activation='relu')(x)
x = Dropout(0.25)(x)
x = Dense(64,activation='relu')(x)
x = Dropout(0.25)(x)
x = Dense(2,activation='softmax')(x)
cnn = Model(inputs=vgg.input,outputs=x)
cnn.summary()
```

```python
from keras.preprocessing.image import load_img
load_img(df.iloc[100]['path'],target_size=(224,224))
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
x = np.array(df['path'])
y = np.array(df['target'])
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from keras.utils import to_categorical
y_train_cate = to_categorical(y_train)
y_test_cate = to_categorical(y_test)
print(y_train[0])
print(y_train_cate[0])
```

```python
cnn.compile(loss='categorical_crossentropy', 
              optimizer='adam', metrics=['accuracy'])
```

```python
import numpy as np
from keras.applications.vgg16 import preprocess_input

for _ in range(5):
    idx = np.random.randint(0,X_train.shape[0],30)
    img_list = []

    #將圖片(X-train)以idx的隨機方式叫出30張
    #要fit前必以array形式來做 故先做np.array()
    #看以甚麼神經元模型就import其搭配的preprocess_input轉檔
    #最後將隨機選出且轉完檔的array形式圖片放入list中
    #最後list的形式還要做np.array成為array的形式

    for p in X_train[idx]:
        img = load_img(p,target_size=(224,224))
        img_np = np.array(img)
        img_np_pre = preprocess_input(img_np)
        img_list.append(img_np_pre)

    xs1 = np.array(img_list)
    ys1 = y_train_cate[idx]

    res = cnn.train_on_batch(xs1,ys1)
    print('Loss:',res)


    import numpy as np
    from keras.applications.vgg16 import preprocess_input
    validx = np.random.randint(0,X_test.shape[0],30)
    val_img_list = []
    for p in X_test[validx]:
        img = load_img(p,target_size=(224,224))
        img_np = np.array(img)
        img_np_pre = preprocess_input(img_np)
        val_img_list.append(img_np_pre)

    xs = np.array(val_img_list)
    ys = y_test_cate[validx]

    res_test=cnn.test_on_batch(xs,ys)
    print('VAL loss:',res_test)
```
