```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from keras.datasets.mnist import load_data
#((train_images,train_answer),(test_images,test_answer))
(x_train,y_train),(x_test,y_test) = load_data()
```

# 圖片處理: MLP 只能以一維處理-->reshape

```python
#W:weight 銓重要有預期區間  a.[0~1] b.[-1~1]  <-故以上的稀疏矩陣範圍([0~255])不相同
#為符合預期輸入 要將x_train做轉換 : 1.rows*columns攤開 2.0~255 > 0~1

x_train_reshape = x_train.reshape(x_train.shape[0],x_train.shape[1]*x_train.shape[2])/255
x_test_reshape = x_test.reshape(x_test.shape[0],x_test.shape[1]*x_test.shape[2])/255

print(x_train_reshape[c])

len(x_train_reshape[c])
```

```python
#從keras做one hot encoding
from keras.utils.np_utils import to_categorical
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

print('原本:',y_train[0])
print('轉化:',y_train_cat[0])
```

```python
from keras.models import Sequential
from keras.layers import Dense

mlp = Sequential()

#第一層:共784層(row*columns => 28*28): Dense設定2的8次方以上表示,老師表示最常用256/128 要注意必須加上 input_dim
#10個actuvation有10個bias, 28*28=784-->784*256+256 = 200960 = Param  <-總共需要訓練的分數(神經元)
#relu是最常用的activation function
mlp.add(Dense(256,activation = 'relu',input_dim = 784))

#最後一層: 共10層(來自答案的y_train/y_test的種類(0~9),且不需要input_dim
#10個actuvation有10個bias, 256*10+256 = Param  <-總共需要訓練的分數(神經元)
mlp.add(Dense(10,activation = 'softmax'))

mlp.summary()
```

```python
mlp1 = Sequential()

#10個actuvation有10個bias, 784*128+128 = 100480 = Param  <-總共需要訓練的分數(神經元)
mlp1.add(Dense(128,activation = 'relu',input_dim = 784))

#最後一層: 共10層(來自答案的y_train(1~10),且不需要input_dim
#10個actuvation有10個bias, 128*10+128 = 1290 = Param  <-總共需要訓練的分數(神經元)
mlp1.add(Dense(10,activation = 'softmax'))

mlp1.summary()
```

```python
mlp.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#batch:看多少筆資料做一次梯度下降(修正參數),拿多少筆來做 -->越小越準-->batch=1就是拿全部做 當然準
#epoch : 整份資料看幾次(有點類似Cross Validation)(決定模型停止的時間點-->要看val_loss的低點且開始平滑來決定epochs次數)
#verbose :印log的程度 1-->進度條 2-->印出必要的 0-->完全不印出
#停止模型訓練: 由欠擬合接近到過擬合 
#條件換算: epoch 10,batch 200: 60000*10/200 = 3000 開始做梯度下降

mlp.fit(x_train_reshape,y_train_cat,
        batch_size=200,epochs=5,
        validation_split=0.1,verbose=2)
```

```python
mlp.evaluate(x_test_reshape,y_test_cat)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
y_pre = mlp.predict_classes(x_test_reshape)
pd.DataFrame(confusion_matrix(y_test,y_pre))
```

```python
#找出所有不相同的值(預測錯誤者)得座標-->原本為tuple形式 故要取出index [0]
idx = np.nonzero(y_pre != y_test)[0]

#取出預測錯誤的image
false_pre_img = x_test[idx]

#取出預測錯誤的 實際值(來自test data)
false_pre_test_label = y_test[idx]

#取出預測錯誤的 預測值(來自predict data)
fasle_pre_pre_label = y_pre[idx]
```

```python
#作圖
width = 10
height = len(fasle_pre_pre_label)//10+1
plt.figure(figsize=(14,40))

for i , img in enumerate(false_pre_img):
    plt.subplot(height,width,i+1)
    t=('[O]:'+str(false_pre_test_label[i])+'[P]:'+str(fasle_pre_pre_label[i]))
    plt.title(t)
    plt.axis('off')
    plt.imshow(img)
```
