# CNN實作 + Multi Layer Perception

```python
#Flatten 將輸入轉成一維
#Conv2D CNN function
#Maxpooling2D 池化
#Sequential序列化 model 但不是神經元本體
from keras.layers import Dense,Flatten
from keras.layers import MaxPooling2D,Conv2D
from keras.models import Sequential
```

```python
cnn = Sequential()

#Conv2D
#filter 過濾器 該曾需要多少過濾器(64,128,256每層不同)
#kernel_size 過濾器大小 --> 3*3 固定即可 -->tuple表示
#strides 每做一次卷積而移動的步長
#padding 設定寬與高的固定與否 ('same'/'valid')
#activation 激活函數 必選relu
#input_dim若有兩個維度改為input_shape
cnn.add(Conv2D(64,(3,3),padding='same',activation='relu',input_shape=(X_train.shape[1],X_train.shape[2],X_train.shape[3])))
#MaxPooling2D
#pool_size -->最大值池化的寬與高 -->default -->*1/2
cnn.add(MaxPooling2D())

#再做一次捲積並池化--> 不用給input因為來自前一層的output ; VGG由兩倍擴充 -->64*2 
cnn.add(Conv2D(128,(3,3),padding='same',activation='relu'))
cnn.add(MaxPooling2D())

#再做一次捲積並池化-->不用給input因為來自前一層的output ; VGG由兩倍擴充 -->128*2
cnn.add(Conv2D(256,(3,3),padding='same',activation='relu'))
cnn.add(MaxPooling2D())

#平坦化:轉為一維(拉平)
cnn.add(Flatten())

#CNN完成後再放回MLP並以Y的數量做輸出層(output)
cnn.add(Dense(512,activation='relu'))
cnn.add(Dense(256,activation='relu'))
cnn.add(Dense(128,activation='relu'))
#輸出: 1. output=10 來自y的種類數量 2. activation 為softmax
cnn.add(Dense(10,activation='softmax'))
cnn.summary()
```

```python
from keras.utils import to_categorical
```

```python
#做正規化 和keras的one hot
X_train_norm = X_train/255
X_test_norm = X_test/255

y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
```

```python
#optimizer = tf.train.AdamOptimizer(learning_rate=0.001).minimize(model.cost)
```

```python
#compile模型 -->代入 loss,optimizer,mertics
cnn.compile(loss='categorical_crossentropy', 
              optimizer='adam', metrics=['accuracy'])
```

```python
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession
config = ConfigProto()
config.gpu_options.allow_growth = True
session = InteractiveSession(config=config)
```

```python
cnn.fit(X_train_norm,y_train_cat,
       batch_size=200,epochs=7,
       validation_split=0.1)
```

```python
y_pre = cnn.predict_classes(X_test_norm)
y_pre
```

```python
#找出所有不相同的值(預測錯誤者)得座標-->原本為tuple形式 故要取出index [0]
idx = np.nonzero(y_pre != y_test.reshape(10000))[0]
idx = idx[:200]
#取出預測錯誤的image
false_pre_img = X_test[idx]

#取出預測錯誤的 實際值(來自test data)
false_pre_test_label = y_test[idx]

#取出預測錯誤的 預測值(來自predict data)
fasle_pre_pre_label = y_pre[idx]
```

```python
#作圖
width = 10
height = len(fasle_pre_pre_label)//10+1
plt.figure(figsize=(14,40))

for i , img in enumerate(false_pre_img):
    plt.subplot(height,width,i+1)
    t=('[O]:'+str(false_pre_test_label[i][0])+'[P]:'+str(fasle_pre_pre_label[i]))
    plt.title(t)
    plt.axis('off')
    plt.imshow(img)
```
